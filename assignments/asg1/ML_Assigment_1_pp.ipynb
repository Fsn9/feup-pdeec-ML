Consider the linear regression problem (with loss function $J(\boldsymbol{\theta}) = \|Y-X\boldsymbol{\theta}\|^2$) for the following data set:

with the model $y = f_{\boldsymbol{\theta}}(x)$, where  $f_{\boldsymbol{\theta}}(x) = \theta_0 + \theta_1 x_1 + \theta_2 \sin(x_1+x_2) + \theta_3 \cos (x_1+x_2)$. 

a) Obtain the expression for the gradient $\nabla J(\boldsymbol{\theta})$ as a function  of the vector of labels $Y$ and the related matrix of features $X$. Obtain explicitly $X$ and $Y$. 
	



$\color{green}{Solution\text{ } 1a)}$  **(Paste a digitalized image of your solution below)**





$J\left(\theta \;\right)={\left\|Y-X\theta \right\|}^2 ={\left(Y-X\theta \right)}^T \left(Y-X\theta \right)=Y^T Y-2\theta^T X^T Y+\theta^T X^T X\theta$
$\Rightarrow \nabla J\left(\theta \right)=\left\lbrack Y^T Y-2\theta^T X^T Y+\theta^T X^T X\theta \right\rbrack d\theta = 2\left(X^T X\theta -X^T Y\right)$

$\mathrm{where}\;X=\left\lbrack \begin{array}{cccc}
\;1 & \frac{\pi }{4} & \sin \left(\frac{\pi }{4}+\frac{\pi }{4}\right) & \cos \left(\frac{\pi }{4}+\frac{\pi }{4}\right)\\

1 & \frac{\pi }{8} & \sin \left(\frac{\pi }{8}+\frac{\pi }{8}\right) & \cos \left(\frac{\pi }{8}+\frac{\pi }{8}\right)
\end{array}\right\rbrack$

$\mathrm{and}\;Y=\left\lbrack \begin{array}{c}
2\\
-2\\
1
\end{array}\right\rbrack \;\;\;\;\;$

b) Say if the following statement is true or false, justifying your response. "The previous linear regression problem has a closed-form solution and the minimization of the loss function has led to an unique solution".

$\color{green}{Solution\text{ } 1b)}$ **(Present your justification here)**


```python
import numpy as np
from math import isclose, pi, sin, cos
import matplotlib.pyplot as plt
from statistics import mean

x = np.array([[1, pi/4, sin(pi/4+pi/4), cos(pi/4 + pi/4)],[1, -pi/3, sin(-pi/3+pi/3), cos(-pi/3 + pi/3)],[1, pi/8, sin(pi/8+pi/8), cos(pi/8 + pi/8)]])
print(x.shape)
print(x)
xx = np.dot(np.transpose(x),x)
print(xx.shape)

# by using the matrix determinant, if det is 0 then the matrix is singular; it is not invertible.
print("det: ", np.linalg.det(xx))
if isclose(np.linalg.det(xx), 0, abs_tol = 1e-10):
    print("false")
else:
    print("true")

# by using single value decomposition, one can verify that since s matrix is positive definitive.
u, s, vh = np.linalg.svd(xx)
print("u: ", u)
print("s: ", s)
print("vh: ", vh)
```

Consider the following data for a linear regression problem where $x_i$ is the feature $i^{th}$ and y is the output:


\begin{array}{llll}
		\hline \hline
		x_1  & x_2 & x_3 & y \\
		\hline \hline
		340    & 16    & 356   & 1.5        \\
		665    & 25    & 690   & 2.8        \\
		368    & 15    & 383   & 1.7        \\
		331    & 15    & 346   & 1.3      \\
		954    & 40    & 994   & 5.0        \\
		\hline
\end{array}
	


Calculate the solutions for the following regression models. If the solution is not unique, please provide a better approach. **Note: do not use the sklearn library**. 

a) $\hat{y} = \theta_1 x_1 + \theta_2 x_2$.

```python

# Function to check the invertability of a matrix by checking the singular values thru Singular Value Decomposition:
def check_if_invertible(matrix):
    u, s, vh = np.linalg.svd(matrix)
    for i in s:
        if isclose(i, 0, abs_tol = 1e-10):
            return False
    return True

# 1. Saving data
x1 = np.array([340, 665, 368, 331, 954]).reshape(-1, 1)
x2 = np.array([16, 25, 15, 15, 40]).reshape(-1, 1)

Y = np.array([1.5, 2.8, 1.7, 1.3, 5.0]).reshape(-1, 1)

# 2. Make inputs X for both a) and b)
Xa = np.concatenate((x1,x2), axis = 1)
Xa2 = np.dot(Xa.T,Xa)

# 3. Find optimal weights: w* = (X'X)^(-1) * X' * Y
# a) and b)
if check_if_invertible(Xa2):
    Wa = np.dot(np.dot(np.linalg.inv(Xa2), Xa.T), Y)
    print(f'Solution for a) is: {Wa}')
    print(f'Predictions for model a) are: {np.dot(Xa, Wa)}')
    x_plot_a = np.concatenate((x1,x2),axis=0)
    y_plot_a = np.concatenate((Y,Y),axis=0)
    plt.scatter(x_plot_a,y_plot_a)
    y_pred_a = np.dot(Xa, Wa)
    y_plot_pred_a = np.concatenate((y_pred_a,y_pred_a),axis=0)
    plt.scatter(x_plot_a,y_plot_pred_a)
    plt.legend(["real values", "predicted values"])
    plt.show()
    MSE_a = (y_plot_a-y_plot_pred_a)**2
    print ("MSE: \n", MSE_a)
else:
    print("Matrix is singular")




```

b) $\hat{y} = \theta_1 x_1 + \theta_2 x_2 + \theta_3 x_3$

```python
x3 = np.array([356, 690, 383, 346, 994]).reshape(-1, 1)

Xb = np.concatenate((x1,x2,x3), axis = 1)
Xb2 = np.dot(Xb.T,Xb)

print("With no regularization: ")
if check_if_invertible(Xb2):
    Wb = np.dot(np.dot(np.linalg.inv(Xb2), Xb.T), Y)
    print(f'Solution for a) is:\n {Wb}')
    print(f'Predictions for model a) are:\n {np.dot(Xb, Wb)}')
    x_plot_b = np.concatenate((x1,x2,x3),axis=0)
    y_plot_b = np.concatenate((Y,Y,Y),axis=0)
    plt.scatter(x_plot_b,y_plot_b)
    y_pred_b = np.dot(Xb, Wb)
    y_plot_pred_b = np.concatenate((y_pred_b,y_pred_b,y_pred_b),axis=0)
    plt.scatter(x_plot_b,y_plot_pred_b)
    plt.legend(["real values", "predicted values"])
    plt.show()
    MSE_b = (y_plot_b-y_plot_pred_b)**2
    print ("MSE: \n", MSE_b)
else:
    print("Matrix is singular")

lambda_values = [1e-6, 1e-3, 1e-2, 0.1, 1, 10, 20, 30, 50, 75, 100]
MSE_sum = []
MSE_avg = []


for lambda_ in lambda_values:
    print("\nregularization with a penalty function lambda =", lambda_)
    Xb2 = np.dot(Xb.T,Xb)+ lambda_*np.eye(np.dot(Xb.T,Xb).shape[0])

    if check_if_invertible(Xb2):
        Wb = np.matmul(np.matmul(np.linalg.inv(Xb2), Xb.T), Y)
        print(f'Solution for theta is:\n {Wb}')
        print(f'Predictions for model b) are:\n {np.dot(Xb, Wb)}')
        x_plot_b = np.concatenate((x1,x2,x3),axis=0)
        y_plot_b = np.concatenate((Y,Y,Y),axis=0)
        plt.scatter(x_plot_b,y_plot_b)
        y_pred_b = np.dot(Xb, Wb)
        y_plot_pred_b = np.concatenate((y_pred_b,y_pred_b,y_pred_b),axis=0)
        plt.scatter(x_plot_b,y_plot_pred_b)
        plt.legend(["real values", "predicted values"])
        plt.title(("Solution for b) with lambda =", lambda_))
        plt.show()
        MSE_b = (y_plot_b-y_plot_pred_b)**2
        print ("MSE: \n", MSE_b)
        MSE_sum.append(sum(MSE_b))
        MSE_avg.append(sum(MSE_b)/len(MSE_b))
    else:
        print("Matrix is singular")

plt.scatter(lambda_values, MSE_sum, edgecolors="black")
plt.title("Relation between MSE sum and different lambda values")
plt.show()

plt.scatter(lambda_values, MSE_avg, edgecolors="black")
plt.title("Relation between avg sum and different lambda values")
plt.show()

```
